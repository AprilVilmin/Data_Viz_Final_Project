```python
# Add the Pandas dependency.
import pandas as pd
```

```python
# Files to load
#File from https://www.kaggle.com/code/thuynyle/hawai-i-s-housing-market-post-covid/data?select=county_market_tracker.tsv000
housing_data_to_load = "Resources/county_market_tracker.tsv"
housing_data_export = "Resources/county_market_tracker_2020_2021.csv"
```

```python
# Read the housing data file and store it in a Pandas DataFrame.
housing_data_df = pd.read_csv(housing_data_to_load, sep='\t')
```

```python
housing_data_df['period_begin']= pd.to_datetime(housing_data_df['period_begin'])
housing_data_df['period_end']= pd.to_datetime(housing_data_df['period_end'])
housing_data_df['last_updated']= pd.to_datetime(housing_data_df['last_updated'])
```

```python
# Determine data types for the housing DataFrame.
housing_data_df.dtypes
```

```python
housing_data_2020_2021_df = housing_data_df.loc[housing_data_df['period_begin'] > '2020-01-01']
housing_data_2020_2021_df
```

```python
housing_data_2020_2021_df.to_csv(housing_data_export, index=False)
```
